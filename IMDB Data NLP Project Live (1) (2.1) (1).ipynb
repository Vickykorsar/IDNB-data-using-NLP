```python
Data=[]
P=PorterStemmer() #for stemming 

for i in range(train.shape[0]):# to get a value range of i from 0 to 75000
    A=train.loc[i,'review'] #to get i'th index of review column data (i'th number review)
    for j in sent_tokenize(A): #to get every sentense seperately from each value(A)to store in j
        B=re.sub('[^A-Za-z0-9]',' ',j)#to remove unwanted panctuations and replace with ' 'space.
        B=B.lower()#lower the alphabeticat words
        B=B.split() #to get every word from each sentenses(j) which are in each value(A) of data of index i
        C=[P.stem(k)  for k in B if k not in set(stopwords.words('english'))]#to remove words which are present in stopwords and tremaining word stould pass to stemming process and save in C
        D=' '.join(C)#to join each word after process with ' 'space to complete the sentense
    Data.append(D)
```

```python
def vectorizer(Token,W2v,vector_size=200): #token is input
    vector=[W2v.wv[i] for i in Token if i in W2v.wv ]#to check the element in vocabulary from single iteration of data rows and then to perform vectorization
    if vector:
        return np.mean(vector,axis=0) #average the values
    else:
        return np.zeros(vector_size) #fill 0 where the words are not there
```

```python
TR['Vectors']=TR['Tokens'].apply(lambda x:vectorizer(x,W2v,vector_size=200))#to check on all iterations one by one here the x is token
TR
```
